Calculate the Moment of a Distribution

```python
#Ques1- a python function to calculate the moment of a distribution. 
# It takes a series and degree as input and output the moment

import pandas as pd
import numpy as np
import math
a=[]
a=[int(i) for i in input ("enter series element: ").split()]

degree = int (input())

def moment(series, degree):
    n= len(series)
    moment = sum(x ** degree for x in series)/n
    return moment

print ("moment of order", degree, "for this series is: ", math.ceil(moment(a,degree)*100)/100)
```

Skewness, Kurtosis, and Jarque-Bera Test

```python
#Ques2 & 3
df=pd.read_csv('csv assignment2.csv')
df
```

```python
#Q2&3

def skewness(x):
    return (moment(x,3)/(np.std(x)**3))
def kurtosis(x):
    return (moment(x,4)/(np.std(x)**4))
def jarque_bera(x):
    s=skewness(x)
    k=kurtosis(x)
    ans=(len(x)/6)*(s**2+((k-3)**2)/4)
    return ans
def jarquebera_result(x):
    if(jarque_bera(x)==0):
        return(True) 
    #because if jarque_bera is 0, s is 0 and k is 3. so normal distribution.
    else:
        return(False)
    
    
for i in df.columns[1:]:
    print('we get', jarquebera_result(df[i]), 'for', i, 'because its test static is', math.ceil(jarque_bera(df[i])*100)/100)
```
